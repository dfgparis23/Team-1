# Opportunities visualisation

```python
opportunities.Stage.value_counts()
```

```python
share_won = opportunities.Stage.value_counts().Won / opportunities.shape[0]
share_lost = opportunities.Stage.value_counts().Lost / opportunities.shape[0]
share_other = 1 - share_won - share_lost
```

```python
opportunities_won = opportunities[opportunities.Stage == "Won"]
opportunities_lost = opportunities[opportunities.Stage == "Lost"]
opportunities_other = opportunities[~((opportunities.Stage == "Lost") | (opportunities.Stage == "Won")).values]
```

```python
opportunities_won
```

```python
plt.figure(figsize=(10,10))
plt.pie([share_won,share_lost,share_other],labels=("Opportunities Won","Opportunities Lost","Other"),autopct='%1.1f%%')
plt.title("Types of different opportunities")
plt.show()
```
